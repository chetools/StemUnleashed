```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
x = np.linspace(-10,10,50)
y= np.random.normal(2.*x+1., scale=2.)
```

```python
A = np.c_[x, np.ones_like(y)]
v = np.linalg.solve(A.T @ A, A.T @ y)
```

```python
v
```

```python
fig=make_subplots()
fig.add_scatter(x=x,y=y, mode='markers')
fig.add_scatter(x=x, y=A@v, mode='lines')
fig.update_layout(width=800, height=600)
```

```python
N=50
theta = np.random.normal(np.linspace(0,2*np.pi, N), 0.03)
r = np.random.normal(3,0.2, N)
x = r*np.cos(theta) + 2.0
y = r*np.sin(theta) + 1.0
```

```python
A=np.c_[2*x, 2*y, np.ones_like(x)]
xc, yc, alpha = np.linalg.solve(A.T @ A, A.T @ (x**2 + y**2))
r = np.sqrt(alpha + xc**2 + yc**2)
```

```python
theta_plot = np.linspace(0,2*np.pi, 100)
x_plot = r*np.cos(theta_plot) + xc
y_plot = r*np.sin(theta_plot) + yc
fig2 = make_subplots()
fig2.add_scatter(x=x,y=y,mode='markers')
fig2.add_scatter(x=x_plot,y=y_plot,mode='lines')
fig2.update_layout(width=600,height=600)
```
