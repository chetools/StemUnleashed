```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
```

```python
oc = np.array([1,1,1])  #origin to center

a_v = np.array([1,1,0.5])  #vector along major axis        
a_v = a_v/np.linalg.norm(a_v)  #normalized vector along major axis

a = 5.  #major radius
b = 3.  #minor radius
c = np.sqrt(a**2 - b**2)  #distance from center to a foci

x=np.linspace(-a,a,30)
y=b*np.sqrt(1-(x/a)**2)
fig=make_subplots()
fig.add_scatter(x=x,y=y,mode='lines')
fig.update_layout(width=600,height=600,template='plotly_dark')

a_vp = np.cross(a_v,a_v+np.array([1,0,0]))  #perturb a_v vector to find a vector normal to a_v
a_vp = a_vp/np.linalg.norm(a_vp) #normalize

vxy = x[:,None]*a_v[None,:] + y[:,None]*a_vp[None,:] #0 slice, 1 xyz

theta = np.linspace(0,2*np.pi,36)

#Rodriguez derivation
vxy_rot = (np.cos(theta)[None,:,None]*vxy[:,None,:] + ((1-np.cos(theta))[None,:,None]) * (((vxy@a_v)[:,None]*a_v[None,:])[:,None,:]) +
           np.sin(theta)[None,:,None]*np.cross(a_v,vxy)[:,None,:])
```

```python
fig = make_subplots(rows=1, cols=1,
                    specs=[[{'type': 'scene'}]])

for circ in vxy_rot:
    fig.add_trace(go.Scatter3d(x=circ[:,0], y=circ[:,1], z=circ[:,2], mode='lines', line_color='rgb(255,255,255)'))

for circ in np.swapaxes(vxy_rot,0,1):
    fig.add_trace(go.Scatter3d(x=circ[:,0], y=circ[:,1], z=circ[:,2], mode='lines', line_color='rgb(255,255,255)'))


fig.update_layout(width=800, template='plotly_dark', showlegend=False)
fig.show()
```
