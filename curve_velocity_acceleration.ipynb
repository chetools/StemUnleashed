```python
import jax.numpy as jnp
import jax
import plotly.figure_factory as ff

import numpy as np
```

```python
def xy(R,theta):
    return jnp.array([R*jnp.cos(theta), R*jnp.sin(theta)])
```

```python
dxy_dtheta=jax.jacobian(xy,(1))
```

```python
d2xy_d2theta=jax.jacobian(dxy_dtheta, (1))
```

```python
xy_vec=jnp.vectorize(xy,signature='(),()->(2)')
dxy_dtheta_vec=jnp.vectorize(dxy_dtheta,signature='(),()->(2)')
d2xy_d2theta_vec=jnp.vectorize(d2xy_d2theta,signature='(),()->(2)')
```

```python
R=1.
theta=jnp.linspace(0,2*np.pi,30)

x,y = xy_vec(R,theta).T
vel_x,vel_y = dxy_dtheta_vec(1., theta).T
a_x,a_y=d2xy_d2theta_vec(1., theta).T
```

```python
fig = ff.create_quiver(x=x, y=y, u=vel_x, v=vel_y, line_color='blue')
fig2 = ff.create_quiver(x=x,y=y,u=a_x,v=a_y, line_color='red')
fig.add_traces(data=fig2.data)
fig.update_layout(width=600,height=600,template='plotly_dark')
```
