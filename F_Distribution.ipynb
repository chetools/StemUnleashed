```python
import numpy as np
from scipy import stats
from plotly.subplots import make_subplots
```

```python
rng=np.random.RandomState(123)
```

```python
N=int(1e6)
dfn=50
dfd=50
num = rng.normal(size=(N,dfn))
den = rng.normal(size=(N,dfd))
samples=np.sum(num**2,axis=1)/np.sum(den**2,axis=1)*dfd/dfn
```

```python
b=10
a=0
n=50
dx=(b-a)/n
counts,bins=np.histogram(samples, np.linspace(a,b,n))
```

```python
xplot=np.linspace(0,np.max(bins),500)
fig=make_subplots()
fig.add_bar(x=(bins[1:]+bins[:-1])/2, y=counts, name='experimental')
fig.add_scatter(x=xplot,y=stats.f.pdf(xplot,dfn,dfd)*dx*N,mode='lines',name='F-distribution')
fig.update_xaxes(range=(0,3),title='value')
fig.update_yaxes(title='counts')
fig.update_layout(width=800,height=600,template='plotly_dark', showlegend=True,
                  title=f'F-Distribution N={N:.1e}, dfn={dfn}, dfd={dfd}')
```
