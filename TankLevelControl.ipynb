```python
import numpy as np
from plotly.subplots import make_subplots
from scipy.integrate import solve_ivp
```

```python
tend=500. #s
q0=0.1 #m3/s
A = 1. #m2
alpha = 0.1  #m2.5/s

h0=(q0/alpha)**2
```

```python
# qin=0.2  #m3/s
def rhs(t,y):
    h=y[0]
    qin = 0.2 + 0.01*np.exp(-t/100.)*np.sin(t/10.)
    return [(qin-alpha*(h**0.5))/A]
```

```python
res=solve_ivp(rhs,(0.,tend), [h0], method='Radau', dense_output=True)
```

```python
tplot=np.linspace(0.,tend,200)
h=res.sol(tplot)[0]
```

```python
fig=make_subplots()
fig.add_scatter(x=tplot,y=h,mode='lines')
fig.update_layout(width=800,template='plotly_dark')
```

```python
A1,A2,A3 = 1., 1., 1.,
alpha = 0.1
q0 = 0.1
h10 = (q0/alpha)**2
h20=h10
h30=h20
tend = 1000.
kc=0.5
taui=100.
taud=-10.
hsp = 1.2*h30
eint0=0.
```

```python
#qmv=kc*(e+ eint/taui + taud*de/dt)
#de/dt = dh3/dt, ignore dhsp/dt spike for step changes in hsp
```

```python
qin=0.1
def rhs(t, y):
    h1, h2, h3, eint = y
    e=hsp-h3
    qmv=kc*(e + eint/taui + alpha*(h2**0.5 - h3**0.5)/A3*taud)
    return [(qin + qmv -alpha*h1**0.5)/A1, alpha*(h1**0.5 - h2**0.5)/A2, alpha*(h2**0.5- h3**0.5)/A3, e]
```

```python
res=solve_ivp(rhs, (0,tend), [h10, h20, h30, eint0], method='Radau', dense_output=True)
```

```python
tplot=np.linspace(0,tend,500)
h1,h2,h3,eint=res.sol(tplot)
```

```python
fig=make_subplots()
fig.add_scatter(x=tplot,y=h1,mode='lines',name='h1')
fig.add_scatter(x=tplot,y=h2,mode='lines',name='h2')
fig.add_scatter(x=tplot,y=h3,mode='lines',name='h3')
fig.update_layout(width=800,template='plotly_dark')
```
