```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
```

```python
a1, a2, a3 = 1., 1., 1.
alpha1, alpha2, alpha3 = 1., 1., 1.
h3sp = 3.

def dh(t,h):
    h1, h2, h3 = h
    q0 = 2*(h3sp-h3)
    q1, q2, q3 = alpha1*np.sqrt(h1), alpha2*np.sqrt(h2), alpha3*np.sqrt(h3)
    return [(q0-q1)/a1, (q1-q2)/a2, (q2-q3)/a3]
```

```python
h_init = [1., 1., 1.]
tend = 100
res=solve_ivp(dh, (0,tend), h_init, method='Radau', dense_output=True)
tplot = np.linspace(0,tend,100)
h1,h2,h3 = res.sol(tplot)
```

```python

fig,ax=plt.subplots(1,3,figsize=(12,4))
ax[0].plot(tplot,h1)
ax[1].plot(tplot,h2)
ax[2].plot(tplot,h3)
```
