```python
import numpy as np
import scipy.stats as stats
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
import pandas as pd
pio.templates.default='plotly_dark'
```

```python
n=1e6
lmbda = 5000
p = lmbda/n
purchase = np.random.choice([0,1], size=int(1e6), p = [(1-p), p])
```

```python
purchase_times = np.where(purchase)[0]
time_between = purchase_times[1:]- purchase_times[:-1]
```

```python
np.histogram(time_between)
```

```python

fig4=px.histogram(time_between, histnorm='probability density')

fig4.update_layout(width=800,height=600)
```
