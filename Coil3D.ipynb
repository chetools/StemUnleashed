```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates_default='plotly_dark'
import numpy as np
```

```python

a, b, c = 1, 0.2, 0.3

u, v = np.mgrid[0:4*np.pi:60j, 0:2*np.pi:18j]

x = (a+c*np.cos(v))*np.cos(u)
y = (a+c*np.cos(v))*np.sin(u)
z = b*u + c*np.sin(v)


fig = make_subplots(rows=1, cols=1,
                    specs=[[{'type': 'scene'}]])

for xs, ys, zs in zip(x,y,z):
    fig.add_trace(go.Scatter3d(x=xs, y=ys, z=zs, mode='lines', line_color='rgb(255,255,255)'))

for xs, ys, zs in zip(x.T,y.T,z.T):
    fig.add_trace(go.Scatter3d(x=xs, y=ys, z=zs, mode='lines', line_color='rgb(255,255,255)'))


fig.add_trace(go.Surface(x=x, y=y, z=z, colorscale=[[0, 'rgb(0,0,255)'], [1,
    'rgb(255,0,0)']], showscale=False))
fig.update_layout(width=800, template='plotly_dark', showlegend=False)
fig.show()
```
