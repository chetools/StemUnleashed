```python
import numpy as np
import scipy.stats as stats
from plotly.subplots import make_subplots
```

```python
rng=np.random.RandomState(123)
p=0.01
N_times=int(1e8)
purchase=rng.choice((0,1),size=(N_times), p=(1-p,p))
purchase_times = np.where(purchase)[0]
```

```python
time_gap=purchase_times[1:]-purchase_times[:-1]
```

```python
N=time_gap.size
```

```python
b=np.max(time_gap)
a=0
n=50
dx=(b-a)/n
counts,bins=np.histogram(time_gap,np.linspace(a,b,n))
```

```python
xplot=np.linspace(0,np.max(bins),500)
fig=make_subplots()
fig.add_bar(x=(bins[1:]+bins[:-1])/2, y=counts, name='experimental')
fig.add_scatter(x=xplot,y=stats.expon.pdf(xplot,scale=1/p)*dx*N,mode='lines',name='exponential-distribution')
fig.update_xaxes(range=(0,400),title='time gap')
fig.update_yaxes(title='counts')
fig.update_layout(width=800,height=600,template='plotly_dark', showlegend=True)
```

```python
period=1000
purchases_in_period = np.sum(purchase.reshape(-1,period),axis=1)
avg = period*p
periods=purchases_in_period.size
```

```python
b=np.max(purchases_in_period)
counts,bins=np.histogram(purchases_in_period,b)
```

```python
counts
```

```python
bins
```

```python
stats.poisson.pmf(bins[:-1],mu=avg)*periods
```

```python

fig=make_subplots()
fig.add_bar(x=bins[:-1], y=counts, name='experimental')
fig.add_scatter(x=bins[:-1],y=stats.poisson.pmf(bins[:-1],mu=avg)*periods,mode='lines',name='poisson-distribution')
fig.update_xaxes(title='purchases over 1000 time units')
fig.update_yaxes(title='counts')
fig.update_layout(width=800,height=600,template='plotly_dark', showlegend=True)
```
