```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
a=0.01
A = 1.
qin_initial_ss = 0.1
sq2g = np.sqrt(2*9.81)
h10 = h20= h30=qin_initial_ss**2/(a**2)/(2*9.81)
tend=5000.
hsp = 1.1*h30
Kc=0.01
taui =200
eint0=0.
```

```python
hsp
```

```python
def rhs(t,v):
    h1,h2,h3, eint =v
    qin=qin_initial_ss
    e=hsp - h3
    qcontrol= Kc*(e + eint/taui)
    q12 = a*sq2g*(h1**0.5)
    q23 = a*sq2g*(h2**0.5)
    q3 = a*sq2g*(h3**0.5)
    return [(qin + qcontrol - q12)/A, (q12 - q23)/A, (q23 - q3)/A, e]
```

```python
res=solve_ivp(rhs, (0.,tend), [h10,h20,h30, eint0], method='Radau', dense_output=True)
tplot = np.linspace(0.,tend,500)
h1,h2,h3,eint=res.sol(tplot)

fig=make_subplots()
fig.add_scatter(x=tplot,y=h1,mode='lines',name='Tank1')
fig.add_scatter(x=tplot,y=h2,mode='lines',name='Tank2')
fig.add_scatter(x=tplot,y=h3,mode='lines',name='Tank3')
fig.update_layout(width=600,height=400,template='plotly_dark')
```
