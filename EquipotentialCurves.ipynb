```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
x,y = np.mgrid[-8:8:101j, -8:8:101j]
```

```python
def d(r1,r2):
    return np.sqrt(np.sum((r1-r2)**2,axis=-1))
```

```python
r = np.stack([x,y],axis=2)
```

```python
v = 1/d(r,np.array([2.5,0])) + 1/d(r,np.array([-2.5,0])) - 1/d(r,np.array([0,2.5])) - 1/d(r,np.array([0,-2.5])) + 1/d(r,np.array([3,3]))
```

```python
cs=[[0,'rgb(255,255,255)'],
    [1,'rgb(255,255,255)']]
```

```python
fig=make_subplots()
fig.add_trace(go.Contour( z=v,x0=-8,dx=0.16,y0=-8,dy=0.16, contours=go.contour.Contours(start=-4, end =4, size=0.1, coloring='lines', 
                                                                                        showlines=True,showlabels=True),
                         colorscale=cs))
fig.update_layout(width=800,height=800, showlegend=False)
fig.update_traces(showscale=False)
```
