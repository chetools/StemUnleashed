# Technical Analysis, trends (lines) in financial data is nonsense baloney

```python
import yfinance as yf
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
import numpy as np
import pandas as pd

# Get stock market data for Nvidia as a Pandas dataframe
data = yf.Ticker("NVDA").history(period='5y')
data
```

```python
# Closing prices look right
fig=px.line(data,y='Close')
fig.update_layout(width=800, template='plotly_dark')
```

```python
# Calculate the daily % change in closing prices and plot a histogram
percent_changes=(data.Close.diff()/data.Close).to_numpy()[1:]
px.histogram(percent_changes, template='plotly_dark' ).update_layout(width=800)
```

```python
# Randomize the order of the daily returns and multiply by closing price on the first day of our data set
# Do you see trends?  Hahaha!
rng=np.random.default_rng()
randomized_returns=rng.permutation(percent_changes)+1.
simulated_prices = np.cumprod(randomized_returns)*data.Close[0]
px.line(y=cumprod,template='plotly_dark').update_layout(width=800)
```

```python
# Lets do this 25 times and see
N=25
simulated_prices = np.cumprod(rng.permutation(np.repeat(percent_changes,N)+1).reshape(-1,N),axis=0)*data.Close[0]
cols = int(np.floor(np.sqrt(N)))
fig2=make_subplots(rows = int(N // cols)+1, cols=cols)
for i in range(N):
    fig2.add_scatter(y=simulated_prices[:,i], row=(i//cols)+1, col=(i%cols)+1)
fig2.update_layout(height=900,width=900, showlegend=False)
```
